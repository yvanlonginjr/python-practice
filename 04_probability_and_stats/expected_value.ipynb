## Problem 1: Dice Game

**Setup:** Roll two fair dice.
- Win **\$10** if the sum is **7**
- Win **\$4** if the sum is **even** (but not 7)
- Lose **\$3** if the sum is **odd** (but not 7)

**Question:** What is the expected value of this game? Should you play?

**Sample space:** $6 \times 6 = 36$ equally likely outcomes.

Count outcomes:

## Problem 2: Coin Flip Game

**Setup:** Flip a fair coin 3 times.
- Win **\$8** if all 3 are heads
- Win **\$2** if exactly 2 heads
- Lose **\$3** if exactly 1 head
- Lose **\$5** if 0 heads

**Question:** What is the fair value to pay to play this game?

**Sample space:** $2^3 = 8$ equally likely outcomes.
Count using $\binom{3}{k}$:

| k (heads) | C(3,k) | Payoff |
|-----------|--------|--------|
| 3 | 1 | +\$8 |
| 2 | 3 | +\$2 |
| 1 | 3 | −\$3 |
| 0 | 1 | −\$5 |

```python
import math


def coin_flip_game_ev():
    total = 2**3  # 8 equally likely outcomes

    cases = [
        (3, math.comb(3, 3),  8,  "All heads"),
        (2, math.comb(3, 2),  2,  "Exactly 2 heads"),
        (1, math.comb(3, 1), -3,  "Exactly 1 head"),
        (0, math.comb(3, 0), -5,  "No heads"),
    ]

    print("Coin Flip Game EV Breakdown")
    print(f"{'Outcome':<20} {'Ways':<6} {'Prob':<10} {'Payoff':<10} {'Contribution'}")
    print("-" * 60)

    ev = 0
    for k, ways, payoff, label in cases:
        prob = ways / total
        contribution = prob * payoff
        ev += contribution
        print(f"{label:<20} {ways:<6} {prob:<10.4f} ${payoff:<9} {contribution:.4f}")

    print("-" * 60)
    print(f"{'TOTAL EV':<20} {total:<6} {'':10} {'':10} ${ev:.4f}")
    print()
    print(f"Fair value to pay: ${ev:.4f}")
    print(f"If entry costs less than ${ev:.2f}, the game has positive EV for you.")
    return ev


coin_flip_game_ev()
```

## Problem 3: Is This Bet Worth Taking?

**Setup:** You're offered a bet on a biased coin.
- The coin lands heads 60% of the time
- You win **\$1.50** on heads
- You lose **\$1.00** on tails
- You can play up to 1,000 times

**Questions:**
1. What is the EV per flip?
2. What is the total expected profit from 1,000 flips?
3. At what win probability does this game become breakeven?

```python
# Biased coin analysis
p_heads = 0.60
p_tails = 1 - p_heads
win     = 1.50
loss    = -1.00
n_flips = 1000

ev_per_flip   = p_heads * win + p_tails * loss
total_ev      = ev_per_flip * n_flips

print(f"EV per flip:      ${ev_per_flip:.4f}")
print(f"Total EV (1,000): ${total_ev:.2f}")

# Breakeven probability: p*1.50 + (1-p)*(-1.00) = 0
# 1.5p - 1 + p = 0 → 2.5p = 1 → p = 0.40
breakeven_p = abs(loss) / (win + abs(loss))
print(f"\nBreakeven win probability: {breakeven_p:.2%}")
print(f"Current win probability:   {p_heads:.2%}")
print(f"Edge over breakeven:       {(p_heads - breakeven_p):.2%}")

# Show EV across different probabilities
print("\nEV scan across win probabilities:")
print(f"{'P(heads)':<15} {'EV/flip':<15} {'Signal'}")
for p in [i/10 for i in range(3, 9)]:
    ev = p * win + (1 - p) * loss
    signal = '✓ TAKE BET' if ev > 0 else '✗ PASS'
    print(f"{p:.0%}{'':<12} ${ev:.4f}{'':<10} {signal}")
```

## Key Takeaways

1. **EV = probability-weighted average payoff** — always decompose into (prob × payoff) terms.
2. **C(n,k) counts favorable outcomes** for symmetric probability spaces (coins, cards).
3. **Breakeven probability** = $\frac{|\text{loss}|}{\text{win} + |\text{loss}|}$ — know this formula.
4. **Positive EV doesn't mean profit every time** — it's a statement about the long-run average.
5. **Enumerate when in doubt** — brute-force verification catches combinatorics errors.
